```python
import pandas as pd
import sys, os
```

```python
for file_name in os.listdir():
    if file_name.endswith(".xls"):
        temp_df = pd.read_excel(file_name)
        save_name = file_name[:-3] + "csv"
        temp_df.to_csv(save_name, index=False)
        print(save_name)
        print(temp_df.head())
```
